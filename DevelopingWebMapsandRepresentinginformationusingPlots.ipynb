```python
import pandas as pd
import os
```

```python
df = pd.read_csv("Volcanoes_USA.txt")
```

```python
df.head()
```

```python
df1 = df[['NAME','ELEV','LAT','LON']]
```

```python
lon = list(df['LON'])
lat = list(df['LAT'])
evel = list(df['ELEV'])
```

```python
print(lon)
print(lat)
print(evel)
```

```python
def getcolor(elevation):
    if elevation<1000:
        return('red')
    elif elevation<=3000:
        return('blue')
    else:
        return('black')
```

```python
for x in evel:
    getcolor(x)
```

```python
map1 = mp.Map(location=[48.111801,-121.111000],zoom_start=3,tiles='stamenterrain')
```

```python
fgv = mp.FeatureGroup(name="Volcanoes")

for lt,ln,el in zip(lat,lon,evel):
    fgv.add_child(mp.Marker(location=[lt,ln],popup=str(el)+" m",icon=mp.Icon(color=getcolor(el))))
map1.add_child(fgv)
map1.add_child(mp.LayerControl())
map1
#map1.save("f.html")
```
