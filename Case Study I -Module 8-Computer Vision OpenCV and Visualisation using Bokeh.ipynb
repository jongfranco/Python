# 1. Write a program to fetch hyperlinks from any website which user enters.

```python
import pandas as pd
import requests
import numpy as np
from bs4 import BeautifulSoup
import re
```

```python
a = input("Enter the website address: ") # i have used https://timesofindia.indiatimes.com/
```

```python
r = requests.get(a)
```

```python
r
```

```python
c = r.content
c
```

```python
soup =BeautifulSoup(c,"html.parser")
soup
```

```python
course = soup.findAll("a",attrs={"href":re.compile("https://")})#,{"class":"toi-widgets clearfix"})
course
# a2=a1.find("a",attrs={"href":re.compile("https://")})
# a2 = a2.get("href")
# a3=a2.get("href")
# a3
```

```python
df = pd.DataFrame(columns=['Link'])
df
```

```python
link1=[]
for i in course:
    link1 = i.get('href')
    if link1!="":
        df1 = pd.DataFrame([[link1]],columns = ['Link'])
        df = df.append(df1)
```
