# 3. Plot a pie-chart of the number of models released by every manufacturer,recorded in the data provide. Also mention the name of the manufacture with the largest releases.

'age': [42, 52, 36, 24, 73],

'postTestScore': [25, 94, 57, 62, 70]

```python
dic = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],'female': [0, 1, 1,0, 1],'age': [42, 52, 36, 24, 73],'preTestScore': [4, 24, 31, 2, 3],'postTestScore': [25, 94, 57, 62, 70]}
```

```python
data = pd.DataFrame(dic)
```

```python
data
```

Draw a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age

```python
sns.scatterplot(x=data['preTestScore'],y=data['postTestScore'],hue=data["age"])
plt.title('Age Wise Test Score Status')
```
