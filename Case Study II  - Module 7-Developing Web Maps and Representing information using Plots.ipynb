```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import folium as mp
from folium import plugins
```

```python
crime = pd.read_csv('Police_Department_Incidents_Year_2016_.csv')
```

# 1. Display first 100 records from the data and play with parameters - tiles, zoom_start etc. Map will be saved in BasicWebMap.html â€“ view it in browser

```python
crime100 = crime.head(100)
```

```python
lon = list(crime100['X'])
lat = list(crime100['Y'])
print(lon)
print(lat)
#evel = list(df['ELEV'])
```

```python
map1 = mp.Map(location=[37.775421,-122.403405],zoom_start=12,tiles='stamenterrain')
```

```python
fgv = mp.FeatureGroup(name="Crime Incidents")

for lt,ln in zip(lat,lon):
    fgv.add_child(mp.Marker(location=[lt,ln],popup="Crime",icon=mp.Icon(color = "red",icon='angry')))#popup=str(el)+" m",icon=mp.Icon(color=getcolor(el)))
map1.add_child(fgv)
map1.add_child(mp.LayerControl())
map1
map1.save("BasicWebMap.html ")
```

# 2. For latest 7 days of data create WebMap of Crimes which are categorized as ROBBERY 

```python
crime1 = crime[crime['Category']=='ROBBERY']
#crime1.head()
```

```python
crime1.sort_values(by=['Date'],ascending=False)
new = crime1['Date'].str.split(" ",n=1,expand=True)
#data["Name"].str.split(" ", n = 1, expand = True) 
crime1['MDY']=new[0]
crime1.dtypes
crime1.sort_values(by=['MDY'],ascending=False)
crime1['MDY'] = pd.to_datetime(crime1['MDY'])
crime1.dtypes
lastdayfrom = pd.to_datetime('12/31/2016')
lastdayfrom
df = crime1.set_index('MDY')
df= df.sort_index()
df = df.loc[lastdayfrom - pd.Timedelta(days=6):lastdayfrom].reset_index()
df['MDY'].unique()
lon1 = list(df['X'])
lat1 = list(df['Y'])
print(lon1)
print(lat1)
#evel = list(df['ELEV'])

Rmap = mp.Map(location=[37.775421,-122.403405],zoom_start=12,tiles='stamenterrain')
fgv = mp.FeatureGroup(name="Robbery Crime Incidents")

for lt1,ln1 in zip(lat,lon):
    fgv.add_child(mp.Marker(location=[lt1,ln1],popup="Robbery Crime",icon=mp.Icon(color = "red",icon='angry')))#popup=str(el)+" m",icon=mp.Icon(color=getcolor(el)))
Rmap.add_child(fgv)
Rmap.add_child(mp.LayerControl())
Rmap
Rmap.save("RobberyMap.html ")
```

# 4. BONUS ASSIGNMENT -- Display heatmap for Divvy Bikes. Divvy Bikes runs bike rental service in Chicago and their bike station geolocation data is shared. Refer to BikesHeatMap.py and Divvy_Stations_2016_Q3.csv, Divvy_Stations_2016_Q4.csv

```python
divvyStations_q3 = pd.read_csv('Divvy_Stations_2016_Q3.csv')
divvyStations_q4 = pd.read_csv('Divvy_Stations_2016_Q4.csv')

# combine and keep the first instance of id
divvyStations = pd.concat([divvyStations_q3, divvyStations_q4], axis=0).drop_duplicates(subset=['id'])
print(divvyStations.head())

CHICAGO_COORD = [41.8781, -87.6298]

map_heat = mp.Map(CHICAGO_COORD,zoom_start=11)

# mark each station as a point
# Uncomment the code below and check

# for index, row in divvyStations.iterrows():
#     folium.CircleMarker([row['latitude'], row['longitude']],
#                         radius=15,
#                         popup=row['name'],
#                         fill_color="#3db7e4",  # divvy color
#                         ).add_to(map_heat)


stationArr = [[row['latitude'],row['longitude']] for index, row in divvyStations.iterrows()]


# plot heatmap
map_heat.add_child(plugins.HeatMap(stationArr, radius=15))
map_heat.save("BikeStationHeatMap.html")
print(map_heat)

```
